```python
import networkx as nx

def create_graph():
    G = nx.Graph()
    
    G.add_nodes_from(range(1,101))
    
    return G
```

```python
def visualize(G,labeldict,n_size):
    #nx.draw(G, labels = labeldict,node_size=n_size)
    nx.draw(G,node_size=n_size)
```

```python
import random
def assign_bmi(G):
    for node in G.nodes():
        G.nodes()[node]['name'] = random.randint(15,40)
        G.nodes()[node]['type'] = 'person'
```

```python
def get_labels(G):
    dict1={}
    for each in G.nodes():
        dict1[each] = G.nodes()[each]['name']
    return dict1
```

```python
def get_size(G):
    array1 =[]
    for each in G.nodes():
        array1.append(G.nodes()[each]['name']*20)
    return array1
```

```python
G=create_graph()
assign_bmi(G)
labeldict=get_labels(G)
nodesize = get_size(G)
visualize(G,labeldict,nodesize)
```
